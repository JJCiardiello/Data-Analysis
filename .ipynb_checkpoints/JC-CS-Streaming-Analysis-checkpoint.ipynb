```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.colors import ListedColormap
import warnings
warnings.filterwarnings("ignore", "use_inf_as_na")
```

Some of the strings have erroneous spaces before/after the text, so the below function trims any strings across the dataframe.

```python
def trimmer(df):
    trim = lambda x: x.strip() if isinstance(x, str) else x
    return df.map(trim)

df = trimmer(df)
```

The below code will fix the date columns, remove any null values and extract month/year values.

```python
df.dropna(subset='date_added', inplace=True)
```

```python
df['date_added'] = pd.to_datetime(df['date_added'])
df['date-year'] = df['date_added'].dt.year.astype(int)
df['date-month'] = df['date_added'].dt.month.astype(int)
df['date-day'] = df['date_added'].dt.strftime('%A')
```

```python
df.head()
```

```python
list = ['rating','duration','country','director','cast']

for i in list:
    if i == 'rating':
        df[i].fillna('Not Rated', inplace=True)
    else:
        df[i].fillna('Unknown', inplace=True)
```

```python
df.info()
```

```python
df['title'].value_counts().sort_values(ascending=False)[df['title'].value_counts().sort_values(ascending=False) > 1]
```

```python
df.drop_duplicates(subset='title', keep='first', inplace=True)
```

```python
df['rating'].unique()
```

```python
df.loc[df['rating'].isin(['74 min','84 min','66 min','NR','UR']),['rating']] = 'Not Rated'
```

```python
df['rating'].unique()
```

```python
df.drop('description', axis=1, inplace=True)
```

```python
df = df.rename(columns={'listed_in': 'categories'})
```

### Data Analysis

Based on our hypothesis, we decided that analysing show genre was an important step in our analysis.

```python
df.groupby(by='categories')[['show_id']].count().sort_values('show_id',ascending=False).head(10)
```

Given that most films contain multiple genres, we made the decision to extract just the first genre and analyse the content based on this.

```python
def split(var):
    return var.split(', ')[0]
```

```python
df['primary-cat'] = df['categories'].apply(split)
```

```python
df_cats = df.groupby(by='primary-cat')[['show_id']].count().sort_values('show_id', ascending=False)
```

```python
fig, axes = plt.subplots(1, figsize=(10,10))
sns.barplot(y=df_cats.index, x=df_cats['show_id'], palette='husl', ax=axes)
axes.set_ylabel('Genre', fontsize=14)
axes.set_xlabel('Shows', fontsize=14)
axes.set_title('Number of shows by genres', fontsize=16, y=1.03)
plt.show()
```

We can from the above bar graph that Netflix produces a large amount of content in the following genres: Drama, Comedy, Action/Adventure, Documentaries, International TV Shows.

We chose to analyse this further in the form of a heatmap so we can see where the top category counts across the years.

```python
df_cat_year = df.groupby(by=['primary-cat','date-year'])[['show_id']].count().sort_values('show_id',ascending=False).unstack()
df_cat_year.fillna(0, inplace=True)
df_cat_year.columns = df_cat_year.columns.droplevel([0])

fig, axes = plt.subplots(1)
sns.heatmap(df_cat_year[[2015,2016,2017,2018,2019,2020,2021]], ax=axes);
axes.set_xlabel('Year', fontsize=12)
axes.set_ylabel('Genre', fontsize=12)
axes.set_title('Heatmap of number of shows by genre and year', fontsize=14, y=1.02)
plt.show()
```

Next we decided to look at the breakdown between movies and TV shows:

```python
cats = df['primary-cat'].unique()
```

```python
df_summ = df.groupby(by='type')['show_id'].count()
df_summ

total = df_summ.values.sum()
```

```python
df_show = df.groupby(by=['date-year','type'])[['show_id']].count().unstack()
df_show.columns = df_show.columns.droplevel([0])
```

```python
fig, axes = plt.subplots(2,1, figsize=(7,7))
axes[0].bar(x=df_show.index, height=df_show['Movie'], color='cornflowerblue', label='Movie')
axes[0].bar(x=df_show.index, height=df_show['TV Show'], color='sandybrown', label='TV Show')
axes[0].set_ylabel('Number', fontsize=12)
axes[0].set_xlabel('Year', fontsize=12)
axes[0].set_title('Number of shows by year', fontsize=13)
axes[0].legend()

axes[1].pie(x=df_summ.values, data=df_summ, labels=df_summ.index, autopct=lambda p: f'{int((p * total)/100+0.5)} ({p:.0f}%)', colors=['cornflowerblue','sandybrown'])
axes[1].set_title('Movie/TV show ratio', fontsize=12, y=1.05)
fig.tight_layout()

plt.show()
```

We can see that Netflix produces more movies than TV shows, and the majority of the production begins in 2015/2016. There is a decrease in content produced from 2020 onwards, but I expect this is due to the COVID-19 pandemic impacting production.

We thought it would be interesting to look at releases across both months and days of the week to see if this will provide any further useful information.

```python
df_month = df.groupby(by='date-month')[['show_id']].count()
df_day = df.groupby(by='date-day')[['show_id']].count()
```

```python
fig, axes = plt.subplots(2,1, figsize=(7,10))

sns.barplot(x=df_month.index, y=df_month['show_id'], ax=axes[0], palette='husl')
axes[0].set_ylabel('Number', fontsize=12)
axes[0].set_xlabel('Month', fontsize=12)
axes[0].set_title('Number of shows by month', fontsize=13)

sns.barplot(x=df_day.index, y=df_day['show_id'], ax=axes[1], palette='husl')
axes[1].set_ylabel('Number', fontsize=12)
axes[1].set_xlabel('Day', fontsize=12)
axes[1].set_title('Number of shows by day', fontsize=13)

fig.tight_layout()

plt.show()
```

There is not any real trend with the month of release, but we can see that the majority of shows are released on a Friday. This will be intentionally done so Netflix can capitalise on people's excitement over the weekend following a release.

Next we chose to look at durations of both movie and TV shows. TV shows are recorded as season, and movies as length in minutes. As the length in minutes is recorded as a string, we need a function to convert this to an integer.

```python
df_dur = df[df['type'] == 'TV Show'].groupby(by='duration')[['show_id']].count().sort_values('show_id',ascending=False)
```

```python
fig, axes = plt.subplots(1, figsize=(10,10))
axes = sns.barplot(x=df_dur['show_id'], y=df_dur.index, palette='husl')
axes.set_ylabel('Duration', fontsize=12)
axes.set_xlabel('Count', fontsize=12)
axes.set_title('Count by season for TV shows', fontsize=14, y=1.03)
plt.show()
```

```python
def duration(var):
    if var != 'Unknown':
        return int(var[:-4])
```

```python
df_movie = df[df['type'] == 'Movie'].copy()
```

```python
df_movie['duration_n'] = df_movie['duration'].apply(duration)
```

```python
fig, axes = plt.subplots(2,1, figsize=(10,10))

sns.histplot(data=df_movie, x='duration_n', bins=100, ax=axes[0], color='cornflowerblue')
axes[0].set_xlabel('Duration, mins', fontsize=13)
axes[0].set_ylabel('Count', fontsize=13)
axes[0].set_title('Histogram of movie length', fontsize=14, y=1.03)

sns.violinplot(data=df_movie, x='duration_n', ax=axes[1], color='sandybrown')
axes[1].set_xlabel('Duration, mins', fontsize=13)
axes[1].set_ylabel('Density', fontsize=13)
axes[1].set_title('Violin plot of movie length', fontsize=14, y=1.03)

plt.tight_layout()
plt.show()
```

We can see that the majority of TV shows on Netflix only last 1 season in length (mode), and movies last around 90-100 mins.

Finally, we decided to look at the countries the content was produced in.

```python
country = df.groupby(by='country')[['show_id']].count().sort_values(by='show_id',ascending=False)[:25]
```

```python
fig, axes = plt.subplots(1)

sns.barplot(data=country, x=country.index, y='show_id', ax=axes)
axes.tick_params(axis='x', rotation=90)
axes.set_xlabel('Country', fontsize=12)
axes.set_ylabel('Count', fontsize=12)
axes.set_title('Shows by country', fontsize=14)
plt.show()
```

We can see that the vast majority of content is produced in the United Stats, India and the U.K.

### Conclusions

1) Confirmed hypothesis that Netflix does cater to a certain segment of the market.
2) Drama, Action/Adventure, International TV Shows/Movies and Comedies are the most popular content.
3) Movies are preferred over TV Shows.
4) Movies on average last 90 mins and TV Shows 1 season.
5) Most content is released on Friday.
6) Heavily produced US content.

As such, we can use this data to drive marketing strategy for CS-Streaming to target the more "untapped" areas of the market. CS-Streaming should:

1) Target genres not typically produced by Netflix.
2) Focus on both TV Shows and Movies, but perhaps produce more TV Shows.
3) Produce some longer-lasting TV shows.
4) Release content towards the end of the week to compete with Netflix.
5) Produce content more globally.

It is worth adding a caveat that the marketing strategy will be heavily dependent on both the budget and timescales to launch CS-Streaming. Targeting the untapped portions of the market will be cheaper/faster to develop, but longer term CS-Streaming will want to push into Netflix's market segment to take some of their market share, and it can do some by slowly adding content that aligns with Netflix's service to steal customers.
